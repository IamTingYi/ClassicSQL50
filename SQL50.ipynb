**<font>28. 統計列印各科成績,各分數段人數:課程ID,課程名稱,verygood[100-86], good[85-71], bad[0-60]</font>**

```python
_sql = [
    "SELECT C.c_id AS '課程ID', C.c_name AS '課程名稱',",
    "SUM(IF(SC.score<=100 AND SC.score >= 86,1,0)) AS 'verygood',",
    "SUM(IF(SC.score<=85 AND SC.score >= 71,1,0)) AS 'good',",
    "SUM(IF(SC.score<=60 AND SC.score >= 0,1,0)) AS 'bad'",
    "FROM sc SC LEFT JOIN course C",
    "ON SC.c_id = C.c_id",
    "GROUP BY C.c_id"
]
dbCursor.execute(' '.join(_sql))
df = pd.DataFrame(list(dbCursor.fetchall()),
                  columns=['課程ID','課程名稱','verygood','good','bad'])
df
```
